#CÓDIGO 1:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pathlib import Path
from sklearn.preprocessing import StandardScaler

# =========================
# FUNCION DESCRIPTORES
# =========================

def metricas(imagen):

    datos = imagen.flatten().astype(np.float64)

    media = np.mean(datos)

    std = np.std(datos)

    energia = np.mean(imagen**2)

    hist = cv2.calcHist([imagen.astype("uint8")],[0],None,[256],[0,256])
    hist = hist/np.sum(hist)
    hist = hist + 1e-7

    entropia = -np.sum(hist*np.log2(hist))

    uniformidad = np.sum(hist**2)

    return media,std,energia,entropia,uniformidad

# =========================
# DESCRIPTOR DE COLOR
# =========================

def metricas_color(imagen):

    histB = cv2.calcHist([imagen],[0],None,[16],[0,256])
    histG = cv2.calcHist([imagen],[1],None,[16],[0,256])
    histR = cv2.calcHist([imagen],[2],None,[16],[0,256])

    histB = histB/np.sum(histB)
    histG = histG/np.sum(histG)
    histR = histR/np.sum(histR)

    return np.concatenate((histB.flatten(),
                           histG.flatten(),
                           histR.flatten()))


# =========================
# EXTRACCION DESCRIPTORES
# =========================

def extraer_descriptores(lista_imagenes,kernel_A,kernel_B,kernel_C):

    descriptores = []

    for imagen in lista_imagenes:

        if imagen is None:
            continue

        # descriptor de color
        color = metricas_color(imagen)

        gray = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
        gray = cv2.equalizeHist(gray)
        gray = cv2.GaussianBlur(gray,(3,3),0)
        gray = cv2.normalize(gray,None,0,255,cv2.NORM_MINMAX)

        F1 = np.abs(cv2.filter2D(gray,cv2.CV_64F,kernel_A))
        F2 = np.abs(cv2.filter2D(gray,cv2.CV_64F,kernel_B))
        F3 = np.abs(cv2.filter2D(gray,cv2.CV_64F,kernel_C))

        d1 = metricas(F1)
        d2 = metricas(F2)
        d3 = metricas(F3)

        vector = np.concatenate((d1,d2,d3,color))

        descriptores.append(vector)

    return np.array(descriptores)


# =========================
# KERNELS
# =========================

kernel_sobel_h = np.array([
[-1,0,1],
[-2,0,2],
[-1,0,1]
],dtype=np.float32)

kernel_sobel_v = np.array([
[-1,-2,-1],
[0,0,0],
[1,2,1]
],dtype=np.float32)

kernel_laplaciano = np.array([
[0,-1,0],
[-1,4,-1],
[0,-1,0]
],dtype=np.float32)
```

#CÓDIGO 6:

```python
def descriptor_individual(imagen,kernel_A,kernel_B,kernel_C):

    color = metricas_color(imagen)

    gray = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray)
    gray = cv2.GaussianBlur(gray,(3,3),0)
    gray = cv2.normalize(gray,None,0,255,cv2.NORM_MINMAX)

    F1 = np.abs(cv2.filter2D(gray,cv2.CV_64F,kernel_A))
    F2 = np.abs(cv2.filter2D(gray,cv2.CV_64F,kernel_B))
    F3 = np.abs(cv2.filter2D(gray,cv2.CV_64F,kernel_C))

    d1 = metricas(F1)
    d2 = metricas(F2)
    d3 = metricas(F3)

    return np.concatenate((d1,d2,d3,color))


def distancia(x,centro):

    x = x.flatten()
    centro = centro.flatten()

    return np.linalg.norm(x-centro)
```

#CÓDIGO 7

```python
carpeta = Path("train")/"american_football"

ruta = carpeta/"A (70).jpg"

imagen_nueva = cv2.imread(str(ruta))

x_nueva = descriptor_individual(
    imagen_nueva,
    kernel_sobel_v,
    kernel_sobel_h,
    kernel_laplaciano
)

x_nueva = scaler.transform([x_nueva])

d_C = distancia(x_nueva,centro_C)
d_S = distancia(x_nueva,centro_S)
d_T = distancia(x_nueva,centro_T)

distancias = [d_C,d_S,d_T]

clases = ["Futbol Americano","Tennis","Hockey"]

indice = np.argmin(distancias)

print("Clasificacion:",clases[indice])
```

#CÓDIGO 8:

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

y_true=[]
y_pred=[]

for i in range(61,101):

    ruta=Path("train")/"american_football"/f"A ({i}).jpg"
    imagen=cv2.imread(str(ruta))

    if imagen is None:
        continue

    x=descriptor_individual(imagen,kernel_sobel_v,kernel_sobel_h,kernel_laplaciano)
    x=scaler.transform([x])

    dist=[distancia(x,centro_C),distancia(x,centro_S),distancia(x,centro_T)]

    pred=np.argmin(dist)

    y_true.append(0)
    y_pred.append(pred)

for i in range(61,101):

    ruta=Path("train")/"baseball"/f"B ({i}).jpg"
    imagen=cv2.imread(str(ruta))

    if imagen is None:
        continue

    x=descriptor_individual(imagen,kernel_sobel_v,kernel_sobel_h,kernel_laplaciano)
    x=scaler.transform([x])

    dist=[distancia(x,centro_C),distancia(x,centro_S),distancia(x,centro_T)]

    pred=np.argmin(dist)

    y_true.append(1)
    y_pred.append(pred)

for i in range(61,101):

    ruta=Path("train")/"basketball"/f"C ({i}).jpg"
    imagen=cv2.imread(str(ruta))

    if imagen is None:
        continue

    x=descriptor_individual(imagen,kernel_sobel_v,kernel_sobel_h,kernel_laplaciano)
    x=scaler.transform([x])

    dist=[distancia(x,centro_C),distancia(x,centro_S),distancia(x,centro_T)]

    pred=np.argmin(dist)

    y_true.append(2)
    y_pred.append(pred)

cm=confusion_matrix(y_true,y_pred)

cm_porcentaje=cm.astype("float")/cm.sum(axis=1)[:,np.newaxis]*100

labels=[]

for i in range(cm.shape[0]):

    fila=[]

    for j in range(cm.shape[1]):

        fila.append(f"{cm[i,j]}\n{cm_porcentaje[i,j]:.1f}%")

    labels.append(fila)

clases=["Futbol Americano","Tennis","Hockey"]

plt.figure(figsize=(6,5))

sns.heatmap(cm,annot=labels,fmt="",cmap="Blues",
xticklabels=clases,
yticklabels=clases)

plt.xlabel("Prediccion")
plt.ylabel("Real")
plt.title("Matriz de Confusion")

plt.show()
```
